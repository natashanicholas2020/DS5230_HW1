B. Use your program to convert the kosarak.dat file to a sparse kosarak.arff. About how long did it take to run?

```python
#converts .dat file into .arff file and records duration
start = time.time()

with open("kosarak.dat", "r") as in_file:
    lines = in_file.readlines()

max_id = max(int(k) for line in lines for k in line.strip().split())

with open("kosarak.arff", "w") as out_file:
    out_file.write("@RELATION news_site_clicks\n")
    for i in range(1, max_id + 1):
        out_file.write(f"@ATTRIBUTE {i} {{0, 1}}\n")
    out_file.write("\n@DATA\n")

    for line in lines:
        st = "{"
        item_ids = sorted(set(int(k) for k in line.strip().split()))
        for item in item_ids:
            st += f"{item - 1} 1,"
        st = st.rstrip(",") + "}"
        out_file.write(st + "\n")

    end = time.time()

    total = end - start

    print(f" {total:.2f} seconds.")
```

It took 0.22s to run.
